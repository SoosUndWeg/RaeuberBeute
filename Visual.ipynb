```python
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.patches as mpatches
import matplotlib.pyplot as plt
from matplotlib.legend_handler import HandlerLine2D
from sympy import* 
init_printing(use_unicode = True )

#Fall 1

def RungeKutta4(r, t, h):                    
       
        k1 = h*f(r, t)
        k2 = h*f(r+0.5*k1, t+0.5*h)
        k3 = h*f(r+0.5*k2, t+0.5*h)
        k4 = h*f(r+k3, t+h)
        return (k1 + 2*k2 + 2*k3 + k4)/6

def f(r, t):
        
        alpha = 0.01
        beta  = 0.25
        delta = 0.05
        gamma = 0.02
        r1    = 0.9
        r2    = 0.8
        x, y = r[0], r[1]
        
        fxd = x * (r1  - alpha * x - gamma * y)
        fyd = y * (-r2 + beta * x - delta * y)
        return np.array([fxd, fyd], float)

h=0.0001                               
tpoints = np.arange(0, 30, h)         
xpoints, ypoints  = [], []
r = np.array([2, 2], float)
for t in tpoints:
        xpoints.append(r[0])          
        ypoints.append(r[1])          
        r += RungeKutta4(r, t, h)  

plt.plot(tpoints, xpoints, color = 'red' ,label = 'Beute')
plt.plot(tpoints, ypoints, color = 'blue',label = 'Räuber')
plt.legend()
plt.xlabel("Zeit")
plt.ylabel("Population")
plt.title("Lotka-Volterra Model")
plt.show()

plt.xlabel("Beute")
plt.ylabel("Jäger")
plt.plot(xpoints, ypoints)
plt.show()

print("Intrinsischen Reproduktions-Sterberaten")
print("Reproduktionsrate = 0.9")
print("Sterberate = -0.8")


print("Interaktionsmatrix:")
Matrix([[-0.01, -0.02], [0.25, -0.05]])

```
